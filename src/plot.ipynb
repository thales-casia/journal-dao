```python
import matplotlib.pyplot as plt
import numpy as np
from math import inf
```

# Break Even

```python
def breakEven(w0:float, w1:float)->int:
  sum:float = 0
  for i in range(1, 1000000):
    sum += 1/(w0 + i * w1)
    if sum >=1:
      return int(i)
  return inf

breakEven(100, 10)
```

```python
w0s = np.array([i * 10 for i in range(5, 16)])
w0s
```

```python
w1s = np.array([i for i in range(1, 11)])
w1s
```

```python
t = np.zeros((len(w0s), len(w1s)))
for i, w0 in enumerate(w0s):
  for j, w1 in enumerate(w1s):
    t[i, j] = breakEven(w0, w1)

t[0, 0]
```

```python
fig, ax = plt.subplots()
im = ax.imshow(t)
ax.set_xticks(np.arange(len(w1s)), labels=w1s)
ax.set_yticks(np.arange(len(w0s)), labels=w0s)
for i, w0 in enumerate(w0s):
  for j, w1 in enumerate(w1s):
    # print(i, j)
    val = t[i, j]
    label = int(val) if val != inf else val
    text = ax.text(j, i, label, ha="center", va="center", color="w")

ax.set_title("Break Even")
fig.set_size_inches(10, 10)
plt.show()
```
